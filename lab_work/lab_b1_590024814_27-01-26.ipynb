## **To display the digits of the 6 digit number**
1. Firstly, we have taken a variable `a` which stores the 6-digit value.  
2. Now, we start the `while` loop which will run till the variable `a` reaches the value of 0.
3. In the loop, we have used the arithmetic operation `%` which gives the remainder when the value is divided by 10,   
which is the last digit of the given number.
4. This value is printed by using the `print` function.
5. Then, we have used the `floor division (//)` to remove the last digit of the number.
6. This process continues till the condition defined in the while loop becomes `false`.

```python
a = 123456
while a!=0:
    print(a%10)
    a=a//10
```

## To check if the given number is palindrome or not
1. Firstly, we have used the variable `a` to store the number which we need to check.
2. We store the same value in variable `b`.
3. Now, we initiate the `while` loop with the condition of `a!=0`.
4. In the loop, we have used the arithmetic operation `%` along with 10 which gives the remiander which is basically, the last value of the digit.
5. This value is then stored in the variable `c`.
6. Since, we need to construct the new digit, we multiply the existing value of c at that point by `10`.
7. Now, we use the `fllor division (//)` to remove the last digit from the number.
8. This continues till the given condition becomes `false`.
9. Now, we use the `if-else` condition to check if the newly constructed digit is same as the provided digit.
10. If it is same, the number is palindrome or else it is not.

```python
a=121
b=a
c=0
while a!=0:
    c=c*10+a%10
    a=a//10
if(c==b):
    print("Palindrome")
else:
    print("Not Palindrome")
print(a, b, c)
```

## To check if the number is armstrong or not
1. Firstly, we have defined a variable `a` to store the value we need to check.
2. We store the same value in other two variables `b` and `d`.
3. Now, we initiate `while` loop with the condition `a!=0`.
4. In this loop, we calculate the number of digits in the given number.
5. Now, we initiate another loop with condition `b!=0`.
6. In this loop, we construct a new number with the condition of armstrong number.
7. Each digit is raised by the variable `count` which stores the number of digit in the number.
8. After doing so, we use the `if-else` condition to check if the newly constructed number is same as the given number.
9. If they are same, then we print that the number is an `armstrong number` or else we print that the number is not an armstrong number.

```python
a=153
b=a
d=a
sum=0
count=0
while a!=0:
    a=a//10
    count+=1
while b!=0:
    c=b%10
    sum+=pow(c, count)
    b=b//10
if(d==sum):
    print("armstorng")
else:
    print("Not Armstrong")
print(sum)
print(count)
```
