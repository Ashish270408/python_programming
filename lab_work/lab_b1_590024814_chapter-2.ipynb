```python
# Not important but good to know

def swap_vowels(s):
    vowels = "aeiouAEIOU"
    chars = list(s)
    
    # find vowel indices
    vowel_indices = [i for i, ch in enumerate(chars) if ch in vowels]
    
    if len(vowel_indices) >= 2:
        first = vowel_indices[0]
        last = vowel_indices[-1]
        chars[first], chars[last] = chars[last], chars[first]
    
    return "".join(chars)

print(swap_vowels("HELLO"))
```

1. **Given:**  
**s = "python is very powerful"**

- **Reverse the order of words.**
- **Reverse each word individually.**
- **Capitalize the first letter of each word.**
- **Count total vowels.**
- **Find the longest word.**


```python
s="python is very powerful"
r_w=[]
words=s.split()
rev_words=words[::-1]
rev_s=" ".join(rev_words)
print("\nRevers order of the words in given sentence is: \n", rev_s)
for i in words:
    rev_w=i[::-1]
    r_w.append(rev_w)
rev_s=" ".join(r_w)
print("\nSentence with each word reversed is: \n", rev_s)
print("\nCapitalizing first letter of each word: \n", s.title())
count=0
for i in s:
    if i in ['a', 'e', 'i', 'o', 'u']:
        count+=1
print("\nTotal number of vowels are: ", count)
long_w=max(words, key=len)
print("\nLongest word is: ", long_w)
```

2. **Given:**  
**s = "aaabbcdddde"**  

- **Find frequency of each character.**
- **Find first repeating character.**
- **Find first non-repeating character.**
- **Compress the string (a3b2c1d4e1).**
- **Remove all duplicate characters.**


```python
s="aaabbcdddde"
freq={}
for i in s:
    freq[i]=freq.get(i, 0)+1
print("\nFrequency of each words is: ", freq)
for i in s:
    if freq[i]>1:
        print("\nFirst repeating character in "+s+" is: ", i)
        break
for i in s:
    if freq[i]==1:
        print("\nFirst non-repeating character in "+s+" is: ", i)
        break
compressed=""
for i in freq:
    compressed+=i
    compressed+=str(freq[i])
print("\nCompressed string is: ", compressed)
unique=""
for i in s:
    if i not in unique:
        unique+=i
print("\nString with only unique character is: ", unique)
```

3. **Given:**  
**nums = [5, 3, 8, 3, 9, 1, 5]**  

- **Remove duplicates (keep order).**

- **Rotate list to the right by 2 positions.**
- **Find all pairs whose sum = 10.**
- **Check if list is sorted.**


```python
nums=[5, 3, 8, 3, 9, 1, 5]
unique=[]
for i in nums:
    if i not in unique:
        unique.append(i)
print("\nList with no duplicates is: ", unique)
largest=second=third=float('-inf')
for i in nums:
    if i>largest:
        third=second
        second=largest
        largest=i
    elif i>second and i!=largest:
        third=second
        second=i
    elif i>third and i!=second:
        third=i
print("\nThird largest num in given list is: ", third)
k=2
rotate=nums[-2:]+nums[:-2]
print("\nNew list with rotated values is: ", rotate)
pairs=[]
target=10
for i in nums:
    complement=target-i
    if i in nums:
        pairs.append((i, complement))
    
print("\nSuch pairs are: ", pairs)
print("\nIs sorted: ", nums==nums.sort())
```
