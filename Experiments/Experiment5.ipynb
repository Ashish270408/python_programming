1. **Write a program to count and display the number of capital letters in a given string.**

```python
a=input("Enter string: ")
count=0
for i in a:
    if i>='A' and i<='Z':
        count+=1
print("Total capital letters in given string", a, "are", count)
```

2. **Count total number of vowels in a given string.**

```python
a=input("Enter string: ")
count=0
for i in a:
    if i=='A' or i=='E' or i=='I' or i=='O' or i=='U' or i=='a' or i=='e' or i=='i' or i=='o' or i=='u':
        count+=1
print("Total vowels in string", a, "are", count)
```

3. **Input a sentence and print words in separate lines.**

```python
a=input("Enter sentence: ")
words=a.split()
for i in words:
    print(i)
```
