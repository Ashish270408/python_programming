1. **Create numpy array to find sum of all elements in an array.**  

```python
import numpy as np

arr = np.array([10, 20, 30, 40, 50])

total_sum = np.sum(arr)

print("Array:", arr)
print("Sum of all elements:", total_sum)
```

2. **Create numpy array of (3,3) dimension. Now find sum of all rows & columns individually. Also find 2nd maximum element in the array.**  

```python
import numpy as np

arr = np.array([[10, 20, 30],
                [40, 50, 60],
                [70, 80, 90]])

print("Array:\n", arr)

row_sum = np.sum(arr, axis=1)
print("\nSum of each row:", row_sum)

col_sum = np.sum(arr, axis=0)
print("Sum of each column:", col_sum)

flat = arr.flatten()
second_max = np.sort(flat)[-2]

print("Second maximum element:", second_max)
```

3. **Perform Matrix multiplication of any 2 n*n matrices.** 

```python
import numpy as np

def get_matrix(n, name):
    matrix = []
    print(f"\nEnter elements of {name} Matrix (each row must have {n} values):")
    
    for i in range(n):
        while True:
            try:
                row = list(map(int, input().split()))
                
                if len(row) != n:
                    print(f"Please enter exactly {n} values!")
                    continue
                
                matrix.append(row)
                break
            except ValueError:
                print("Enter valid numbers only!")
    
    return np.array(matrix)

n = int(input("Enter size of matrix (n x n): "))

A = get_matrix(n, "First")
B = get_matrix(n, "Second")

result = A @ B

print("\nFirst Matrix:\n", A)
print("\nSecond Matrix:\n", B)
print("\nResult of Multiplication:\n", result)
```

4. **Write a Pandas program to get the powers of an array values element-wise.**   
**Note: First array elements raised to powers from second array**  
**Sample data: {'X':[78,85,96,80,86], 'Y':[84,94,89,83,86],'Z':[86,97,96,72,83]}**  
**Expected Output:**  
**X Y Z**  
**0 78 84 86**  
**1 85 94 97**  
**2 96 89 96**  
**3 80 83 72**  
**4 86 86 83**  

```python
import pandas as pd

data = {'X':[78,85,96,80,86],
        'Y':[84,94,89,83,86],
        'Z':[86,97,96,72,83]}

df = pd.DataFrame(data)

print("Original DataFrame:\n")
print(df)

first_row = df.iloc[0]

result = df.pow(first_row)

print("\nResult after applying power:\n")
print(result)
```

6. **Write a Pandas program to find and replace the missing values in a given DataFrame which do not have any valuable information.** 

```python
import pandas as pd
import numpy as np

data = {
    'Name': ['Aman', 'Riya', 'Rahul', 'Sneha', 'Karan'],
    'Marks': [85, np.nan, 78, np.nan, 90],
    'City': ['Delhi', 'Mumbai', None, 'Pune', None]
}

df = pd.DataFrame(data)

print("Original DataFrame:\n")
print(df)

print("\nMissing values in DataFrame:\n")
print(df.isnull())

df['Marks'].fillna(df['Marks'].mean(), inplace=True)

df['City'].fillna("Unknown", inplace=True)

print("\nDataFrame after replacing missing values:\n")
print(df)
```

7. **Create a program to demonstrate different visual forms using Matplotlib.** 

```python
import matplotlib.pyplot as plt

subjects = ['Physics', 'Chemistry', 'Maths', 'English']
marks = [85, 78, 92, 74]

# 1 Line Graph
plt.figure()
plt.plot(subjects, marks, marker='o')
plt.title("Line Graph - Marks")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()

# 2 Bar Graph
plt.figure()
plt.bar(subjects, marks)
plt.title("Bar Graph - Marks")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()

# 3 Scatter Plot
plt.figure()
plt.scatter(subjects, marks)
plt.title("Scatter Plot - Marks")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()

# 4 Pie Chart
plt.figure()
plt.pie(marks, labels=subjects, autopct='%1.1f%%')
plt.title("Pie Chart - Marks Distribution")
plt.show()
```
