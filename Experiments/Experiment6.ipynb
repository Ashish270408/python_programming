1. **Scan n values in range 0-3 and print the number of times each value has occurred.**

```python
l=list(map(int, input("Enter elements: ").split()))
dict={0:0, 1:0, 2:0, 3:0}
for i in l:
        dict[i]+=1
for k in dict:
        print(k, "occurred", dict[k], "times")
```

2. **Create a tuple to store n numeric values and find the average of all values.**

```python
t=tuple(map(int, input("Enter elements: ").split()))
sum=0
for i in t:
    sum+=i
avg=sum/len(t)
print(avg)
```

3. **WAP to input a list of scores for N students in a list data type. Find the score of the runner-up and print the output.**  
**Sample Input**  
**N = 5**  
**Scores= 2 3 6 6 5**  
**Sample output**  
**5**  
***Note*: Given list is [2, 3, 6, 6, 5]. The maximum score is 6, second maximum is 5. Hence, we print 5 as the runner-up score.** 

```python
l=list(map(int, input("Enter Elements: ").split()))
l=sorted(l, reverse=True)
max=l[0]
for i in l:
    if i<max:
        sec_max=i
        break
print("Runner up score is", sec_max)
```

4. **Create a dictionary of n persons where key is name and value is city.**   
**a. Display all names**  
**b. Display all city names**  
**c. Display student names and cities of all students.**  
**d. Count the number of students in each city.**  

```python
n=int(input("Number of names u want to enter: "))
dict={}
for _ in range(n):
    name=input("Enter name: ")
    city=input("Enter city: ")
    dict[name]=city
print("\nAll names:")
for i in dict.keys():
    print(i)
print("\nAll city names: ")
for i in dict.values():
    print(i)
print("\nWho lives where?")
for i, j in dict.items():
    print(i, "lives in", j)
count={}
for city in dict.values():
    count[city]=count.get(city, 0)+1
print("\nHow many people lives win which city?")
for city, person in count.items():
    print(person, "people lives in", city)
```

5. **Store details of n movies in a dictionary by taking input from the user. Each movie must store details like name,  year, director name, production cost, collection made (earning) & perform the following :-**  
**a. print all movie details**  
**b. display name of movies released before 2015**  
**c. print movies that made a profit.**  
**d. print movies directed by a particular director.** 

```python
n=int(input("Number of movies you wish to enter: "))
movies={}
for _ in range(n):
    name=input("Enter name: ")
    year=int(input("Enter the year: "))
    dir_name=input("Enter director name: ")
    cost=int(input("Enter the cost of production: "))
    col=int(input("Enter the collection made: "))
    
    movies[name]={'year':year, 'dir_name':dir_name, 'cost':cost, 'col':col}
print("\nAll movie details: ")
for mov, details in movies.items():
    print(mov, ":", details)
print("\nMovies made before 2015: ")
for name, details in movies.items():
    if details['year']<2015:
        print(name)
print("\nMovies that made profit: ")
for name, details in movies.items():
    if details['col']>details['cost']:
        print(name)
dir=input("Enter the name of director you wish to see movies of: ")
print("\nMovies made by", dir)
for name, details in movies.items():
    if details['dir_name']==dir:
        print(name)
```
