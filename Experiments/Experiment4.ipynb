1. **Find a factorial of given number.**

```python
a=int(input("Enter the number: "))
fact=1
for i in range(1, a+1):
    fact*=i
print(fact)
```

2. **Find whether the given number is Armstrong number.**

```python
a=int(input("Enter a number: "))
b=a
sum=0
count=len(str(a))
while a>0:
    n=a%10
    sum+=pow(n, count)
    a=a//10
if sum==b:
    print("The number is an armstrong number!!!")
else:
    print("The number is not an armstrong number!!!")
```

3. **Print Fibonacci series up to given term.**

```python
n=int(input("Enter the number: "))
a=0
b=1
while n>0:
    print(a)
    c=a+b
    a=b
    b=c
    n-=1
```

4. **Write a program to find if given number is prime number or not.**

```python
a=int(input("Enter a number:"))
n=a
count=0
while n>1:
    if n==a:
        n-=1
        continue
    if a%n==0:
        count+=1
        break
    n-=1
if count:
    print("Not a prime number!!")
else:
    print("Prime number")
```

5. **Check whether given number is palindrome or not.**

```python
a=int(input("Enter a number: "))
b=a
sum=0
while a>0:
    sum=sum*10+a%10
    a=a//10
if sum==b:
    print("The number is a palindrome")
else:
    print("The number is not a palindrome")
```

6. **Write a program to print sum of digits.**

```python
a=int(input("Enter number: "))
sum=0
while a>0:
    sum+=a%10
    a//=10
print(sum)
```

7. **Count and print all numbers divisible by 5 or 7 between 1 to 100.**

```python
count=0
a=1
while a>=1 and a<=100:
    if(a%5==0 or a%7==0):
        print(a)
        count+=1
    a+=1
print("Total numbers: ", count)
```

9. **Print all prime numbers between 1 and 100.**

```python
for i in range (2, 101):
    prime=True
    for j in range (2, i):
        if i%j==0:
            prime=False
            break
    if prime:
        print(i)
```
