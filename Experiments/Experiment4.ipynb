# **Experiment 4: Use of Loops** 
1. **Find a factorial of given number.**

```python
a=int(input("Enter the number: "))
fact=1
while a>0:
    fact*=a
    a-=1
print(fact)
```

2. **Find whether the given number is Armstrong number.**

```python
a=int(input("Enter a number: "))
b=a
sum=0
count=len(str(a))
while a>0:
    n=a%10
    sum+=pow(n, count)
    a=a//10
if sum==b:
    print("The number is an armstrong number!!!")
else:
    print("The number is not an armstrong number!!!")
```

3. **Print Fibonacci series up to given term.**

```python
n=int(input("Enter the number: "))
a=0
b=1
while n>0:
    print(a)
    c=a+b
    a=b
    b=c
    n-=1
```

4. **Write a program to find if given number is prime number or not.**

```python
a=int(input("Enter a number:"))
n=a
count=0
while n>1:
    if n==a:
        n-=1
        continue
    if a%n==0:
        count+=1
        break
    n-=1
if count:
    print("Not a prime number!!")
else:
    print("Prime number")
```
