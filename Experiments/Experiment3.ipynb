1. **Check whether a given number is divisible by 3 and 5 both.**

```python
a=int(input("Enter a number"))
if (a%3 == 0 and a%5 == 0):
    print("The number is divisible by 3 and 5 both")
else:
    print("The number is not divisible by 3 and 5 both")
```

2. **Check whether a given number is multiple of 5 or not.**

```python
a=int(input("Enter a number"))
if(a%5==0):
    print("The number is multiple of 5")
else:
    print("The number is not a multiple of 5")
```

3. **Find the greatest among two numbers. If numbers are equal than print “numbers are equal”.**

```python
a=int(input("Enter first number: "))
b=int(input("Enter second number: "))
if(a>b):
    print("The number", a, "is greater among the two")
elif(b>a):
    print("The number", b, "is greater among the two")
else:
    print("Both the numbers are equal!!")
```
