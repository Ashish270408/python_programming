1. **Check whether a given number is divisible by 3 and 5 both.**

```python
a=int(input("Enter a number"))
if (a%3 == 0 and a%5 == 0):
    print("The number is divisible by 3 and 5 both")
else:
    print("The number is not divisible by 3 and 5 both")
```

2. **Check whether a given number is multiple of 5 or not.**

```python
a=int(input("Enter a number"))
if(a%5==0):
    print("The number is multiple of 5")
else:
    print("The number is not a multiple of 5")
```

3. **Find the greatest among two numbers. If numbers are equal than print “numbers are equal”.**

```python
a=int(input("Enter first number: "))
b=int(input("Enter second number: "))
if(a>b):
    print("The number", a, "is greater among the two")
elif(b>a):
    print("The number", b, "is greater among the two")
else:
    print("Both the numbers are equal!!")
```

4. **Find the greatest among three numbers assuming no two values are same.**

```python
a=int(input("Enter first number: "))
b=int(input("Enter second number: "))
c=int(input("Enter third number: "))
if(a>b and a>c):
    print(a, "is greatest among all")
elif(b>a and b>c):
    print(b, "is greatest among all")
else:
    print(c, "is greatest amonng all")
```

6. **Find whether a given year is a leap year or not.**

```python
a=int(input("Enter the year:"))
if(a%400==0 or (a%4==0 and a%100!=0)):
    print("The year is leap year!!")
else:
    print("The year is not a leap year!!!")
```

8. **Print the grade sheet of a student for the given range of cgpa. Scan marks of five subjects and calculate the percentage.**    
**CGPA=percentage/10**  
**CGPA range:**  
**0 to 3.4 -> F**  
**3.5 to 5.0->C+**  
**5.1 to 6->B**  
**6.1 to 7-> B+**  
**7.1 to 8-> A**  
**8.1 to 9->A+**  
**9.1 to 10-> O (Outstanding)** 

```python
m1, m2, m3, m4, m5=map(int, input("Enter marks of 5 subjects separated by space:").split())
per=((m1+m2+m3+m4+m5)/500)*100
cgpa=per/10
if(cgpa>=0 and cgpa<=3.4):
    print("F")
elif(cgpa>=3.5 and cgpa<=5.0):
    print("C+")
elif(cgpa>=5.1 and cgpa<=6.0):
    print("B")
elif(cgpa>=6.1 and cgpa<=7.0):
    print("B+")
elif(cgpa>=7.1 and cgpa<=8.0):
    print("A")
elif(cgpa>=8.1 and cgpa<=9.0):
    print("A+")
else:
    print("O")
print(per, cgpa)
```
