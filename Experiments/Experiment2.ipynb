3. **Write a Python program to solve (x+y)*(x+y)**  

```python
x, y = map(int, input("Enter value of x and y separated by space: ").split())
ans=(x+y)*(x+y)
print(ans)
```

```python
l = 4
b = 3
h_s = pow(l, 2) + pow(b, 2)
ans = pow(h_s, 0.5)
print(ans)
```
