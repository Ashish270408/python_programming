1. **Write a Python function to find the maximum and minimum numbers from a sequence of numbers.**  
**(Note: Do not use built-in functions.)**

```python
def maximum(list):
    max_value=list[0]
    for i in list:
        if max_value<i:
            max_value=i
    return max_value

def minimum(list):
    min_value=list[0]
    for i in list:
        if min_value>i:
            min_value=i
    return min_value

l1=[2, 3, 1, 0, -12, 56, 42]
print(maximum(l1))
print(minimum(l1))
```

2. **Write a Python function that takes a positive integer and returns the sum of the cube of all the positive integers smaller than the specified number.**

```python
def cube(num):
    sum=0
    num-=1
    while num>0:
        sum+=num**3
        num-=1
    return sum

print(cube(5))
print(cube(4))
```

3. **Write a Python function to print 1 to n using recursion. (Note: Do not use loop)**

```python
def rec_num(num):
    if num==0:
        return
    print(num)
    rec_num(num-1)
    
rec_num(5)
```

4. **Write a recursive function to print Fibonacci series upto n terms.**

```python
def fib(n):
    if n<=1:
        return n
    return fib(n-1)+fib(n-2)

def print_fib(n):
    for i in range(n):
        print(fib(i), end=" ")
        
print_fib(5)
```

5. **Write a lambda function to find volume of cone.**

6. **Write a lambda function which gives tuple of max and min from a list.**    
**Sample input: [10, 6, 8, 90, 12, 56]**  
**Sample output: (90,6)**  

```python
find_max_min = lambda lst: (max(lst), min(lst))
list=[10, 6, 8, 90, 12, 56]
print(find_max_min(list))
```

7. **Write functions to explain mentioned concepts:**  
- **Keyword argument**  
- **Default argument**  
- **Variable length argument** 

```python
def intro (name, age):
    print("Name: ", name)
    print("Age: ", age)

intro("Ashish", 17)
print("This is an Example of keyword argument...\n\n\n")

def say(name, msg="Hello"):
    print(msg+" "+name)
say("Ashish", "Bye")
say("Ashish ")
print("This is an example of default argument...\n\n\n")


def inf(*args):
    sum=0
    for i in args:
        sum+=i
    print(sum)

inf(1, 2, 3, 4)
print("This is an example of variable length argument...")
```
