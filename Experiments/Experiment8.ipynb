1. **Add few names, one name in each row, in “name.txt file”.**    
- **Count no of names**  
- **Count all names starting with vowel**  
- **Find longest name**

```python
with open("names.txt", "w") as f:
    f.write("Ashish\n")
    f.write("Tanveer\n")
    f.write("Bhavya\n")
    f.write("Aryan\n")
    f.write("Vaanya\n")
    f.write("Ritika\n")

with open("names.txt", "r") as f:
    names=f.read().splitlines()
    count=len(names)
    vowel=0
    for i in names:
        if i[0] in "AEIOUaeiou":
            vowel+=1
    long_name=str(0)
    for name in names:
        if len(name)>len(long_name):
            long_name=name

print("Number of names: ", count)
print("Number of names starting with a vowel: ", vowel)
print("Longest name: ", long_name)
```

2. **Store integers in a file.**  
- **Find the max number**  
- **Find average of all numbers**  
- **Count number of numbers greater than 100**  

```python
with open("integers.txt", "w") as f:
    f.write("1 27 13 248 500")

with open("integers.txt", "r") as f:
    numbers=list(map(int, f.read().split()))
    max_number=numbers[0]
    sum=0
    count=0
    for i in numbers:
        if max_number<i:
            max_number=i
        sum+=i
        if i>100:
            count+=1

avg=sum/len(numbers)
print("Maximum number is: ", max_number)
print("Average of all numbers is: ", avg)
print("Numbers greater then 100: ", count)
```

4. **Input two values from user where the first line contains N, the number of test cases. The next N lines contain the space separated values of a and b. Perform integer division and print a/b. Handle exception in case of ZeroDivisionError or ValueError.**  
**Sample input**
- **0**  
- **$**  
- **1**  
**Sample Output:**  
- **Error Code: integer division or modulo by zero**  
- **Error Code: invalid literal for int() with base 10: '$' 3** 

```python
n=int(input())
for _ in range(n):
    try:
        a, b = map(int, input("Enter the values a/b: ").split())
        print(a//b)
    
    except ZeroDivisionError as e:
        print("Error: ", e)
    
    except ValueError as e:
        print("Error: ", e)
```

5. **Create multiple suitable exceptions for a file handling program.**

```python
try:
    filename=input("Enter filename: ")
    with open(filename, "r") as f:
        content=f.read()
        if not content:
            raise ValueError("File is empty")

except FileNotFoundError as e:
    print("Error: ", e)

except PermissionError as e:
    print("Error: ", e)
    
except Exception:
    print("Unexpected error occured")
```
