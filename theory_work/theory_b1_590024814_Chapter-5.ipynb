```python
import json

class JSONWrapper:
    def __init__(self, data: dict):
        if not isinstance(data, dict):
            raise TypeError("Input must be a dictionary")
        self.data=data
    
    def to_json(self):
        return json.dumps(self.data)
    
    def to_pretty_json(self):
        return json.dumps(self.data, indent=4)

data={"name":"Ashish", "marks":{"Physics":20, "Chemistry":32}}
wrapper=JSONWrapper(data)

json_output=wrapper.to_json()
print(json_output)

pretty_output=wrapper.to_pretty_json()
print(pretty_output)
```

```python
class Student:
    def __init__(self):
        self.name=0
        self.roll=0
    def set_details(self, name, roll):
        self.name=name
        self.roll=roll
    
    
    def display(self):
        print(self.name, self.roll)
s1=Student()
s1.set_details("Ashish", 27)
s1.display()
```

```python
import pandas as pd
freq={}

with open("table.txt", "w") as f:
    f.write("Name, Address\n")
    f.write("Ashish, Firozabad\n")
    f.write("Tanveer, Ambala\n")
    f.write("Vaanya, Jalandhar\n")
    f.write("Kulraj, Ambala\n")

with open("table.txt", "r") as f:
    data=f.read().splitlines()
for i in data:
    ind_data=i.split(",")
    name, city = ind_data
    freq[city]=freq.get(city, 0)+1

cities=[]
population=[]
for i in freq:
    cities.append(i)
    population.append(freq[i])
cities.pop(0)
population.pop(0)

table=pd.DataFrame({"City":cities, "Population":population})
print(table)
    
```
