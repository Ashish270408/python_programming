Q1. The function should return the top k elements based on score.

```python
def top_k(items, k):
    items.sort(key=lambda x: x[1], reverse=True)
    return items[:k]
items = [("A", 90), ("B", 95), ("C", 95), ("D", 85)]
k = 2
print(top_k(items, k))
```

Q2. Count occurrences of words under each category.

```python
def count_by_category(pairs):
    d={}
    for cat, word in pairs:
        d.setdefault(cat, {})
        d[cat].setdefault(word, 0)
        d[cat][word]+=1
    return d
pairs = [("sports","win"), ("sports","win"), ("tech","ai")]
print(count_by_category(pairs))
```

Q3. You are given n distinct numbers from range 0 to n. Return the missing number.

```python
def missing_number(nums):
    n=len(nums)
    total=n*(n+1)/2
    return int(total-sum(nums))
nums = [0, 1, 3]
print(missing_number(nums))
```

Q4. The function should:  

- Return a list of unique values added.
- If no second argument provided, each call should start fresh. 

```python
def add_unique(x, seen=set()):
    current=set()
    current.clear()
    if x not in seen:
        seen.add(x)
        current.add(x)
    return list(current)
print(add_unique(1))
print(add_unique(2))
```

Q5. Moving Average of Last w Elements

```python
def moving_average(stream, w):
    result=[]
    window_sum=0
    for i in range(len(stream)):
        window_sum+=stream[i]
        
        if i>=w:
            window_sum-=stream[i-w]
        
        current_size=min(i+1, w)
        result.append(window_sum/current_size)
    return result
print(moving_average([1, 2, 3, 4, 5], 3))
```

Q6. Run-Length Decode with Escape Support

```python
def rle_decode(s):
    result = []
    i=0
    n=len(s)
    while i<n:
        if not s[i].isdigit():
            raise ValueError("Invalid Format")
        count=0
        while i<n and s[i].isdigit():
            count=count*10+int(s[i])
            i+=1
        if i>=n:
            raise ValueError("Invalid Format")
        if s[i]=='\\':
            i+=1
            if i>=n:
                raise ValueError("Invalid Format")
        char=s[i]
        result.append(char*count)
        i+=1
    return "".join(result)
    
rle_decode("3a2b1c")
```

Q7. Longest Unique Substring

```python
def longest_unique_substring(s):
    seen={}
    left=0
    max_length=0
    for right in range(len(s)):
        char=s[right]
        if char in seen and seen[char]>=left:
            left=seen[char]+1
            
        seen[char]=right
        max_length=max(max_length, right-left+1)
        
    return max_length

str="abcabcbb"
print(longest_unique_substring(str))
```
