```python
def normalize_text(s):
    s=s.strip().lower()
    clean=""
    for i in s:
        if i.isalnum() or i.isspace():
            clean+=i
    return clean.split()
a="How are you doing?"
print(normalize_text(a))
```

```python
def normalize_all(l1):
    out=[]
    for s in l1:
        out.append(normalize_text(s))
    return out
sentences = [" Hi, there! ", "Python is FUN."]
print(normalize_all(sentences))
```

```python
def word_count(s):
    freq={}
    words=normalize_text(s)
    for i in words:
        freq[i]=freq.get(i, 0)+1
    return freq

sen="Python, python! is easy; Python is fun."
print(word_count(sen))
    
```
