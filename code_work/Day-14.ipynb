1. **Create a list of 10 integers.**  
**Then:**  
- **Print the first and last element.**
- **Replace the element at index 3 with 100.**
- **Reverse the list without using reverse().**
- **Remove the 5th element.**
- **Find the second largest number without using sort().**



````python
a=list(range(1, 11))
print("\nThe list is: ", a)
print("\nThe first element is", a[0], "and the last element is ", a[-1])
a[3]=100
print("\nThe new list is: ", a)
rev=a[::-1]
print("\nThe reversed list is: ", rev)
a.pop(4)
print("\nThe new list after removing is: ", a)
largest=0
second=largest
for num in a:
    if num>largest:
        second=largest
        largest=num
    elif num>second and num!=largest:
        second=num
print("\nThe second largest element is: ", second)
````

2. **Given the list:**  
**numbers = [4, 7, 2, 7, 9, 2, 7, 5, 2]**
- **Count how many times each number appears (without using collections).**
- **Find the most frequent number.**
- **Remove duplicates while keeping order.**
- **Create a new list containing only even numbers.**
- **Check if the list is a palindrome.**


````python
a=[4, 7, 2, 7, 9, 2, 7, 5, 2]
freq={}
for i in a:
    freq[i]=freq.get(i, 0)+1
print("\nEach number appeard following times: ", freq)
most=max(freq, key=freq.get)
print("\nThe number with the maximum freq is: ", most)
unique=[]
for i in a:
    if i not in unique:
        unique.append(i)
print("\nList with unique elements is: ", unique)
even=[]
for i in a:
    if i%2==0:
        even.append(i)
print("\nThe list with even numbers is: ", even)
print("\nIs palindrome: ", a==a[::-1])
````

3. **Create a dictionary representing a student:**
```python
{
    "name": "Rahul",
    "age": 20,
    "marks": [78, 85, 90]
}
```
**Then:**

- **Add a new key "grade" with value "A".**
- **Update age to 21.**
- **Add a new mark to the marks list.**
- **Calculate the average marks.**
- **Remove the "grade" key.**


````python
dict={"name":"Rahul", "age":20, "marks":[78, 85, 90]}
dict["grade"]="A"
print("\nThe new dictionary is: ", dict)
dict["age"]=21
print("\nThe new dictionary is: ", dict)
dict["marks"].append(95)
print("\nThe new dictionary is: ", dict)
avg=np.sum(dict["marks"])/len(dict["marks"])
print("\nThe average is: ", avg)
dict.pop("grade")
print("\nThe new dictionary is: ", dict)
````
