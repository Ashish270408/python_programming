1. **Create a function called greet().**  

- **It should take one parameter: name.**
- **It should print: "Hello, <name>!"**
- **Call the function with your name.**
- **Modify the function so that it returns the greeting instead of printing it.**


```python
def greet(name):
    print("Hello,", name, "!")
greet("Ashish")
def greet2(name):
    return "Hello, "+name+" !"
greetings=greet2("Ashish")
print("\n", greetings)
```

2. **Create a function called check_even().**  

- **It should take one number as input.**
- **It should return True if number is even.**
- **It should return False if number is odd.**
- **Call the function with different numbers.**
- **Use the function inside an if statement.**


```python
def check_even(num):
    if num%2==0:
        return True
    else:
        return False
print(check_even(22))
print(check_even(25))
if check_even(23):
    print("Even number")
else:
    print("Odd number")
```

3. **Create a function called find_max().**  

- **It should take a list as input.**
- **It should return the largest number.**
- **Do NOT use max().**
- **Handle empty list (print message or return None).**
- **Call the function with at least two different lists.**


```python
def find_max(list):
    if len(list)==0:
        print("\nList is Empty!!!")
        return None
    largest=list[0]
    for i in list:
        if i>largest:
            largest=i
    return largest
print(find_max([2, 4, 9, 1, 45, 87, 2, 81]))
print(find_max([]))
```
