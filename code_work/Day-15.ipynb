1. **Given a string:**  
**s = "cybersecurity"**  

- **Print the first and last character.**
- **Print characters from index 2 to 6.**
- **Reverse the string without using reversed().**
- **Count how many times letter 'e' appears.**
- **Convert the string to uppercase.**


```python
s="cybersecurity"
print("\nThe first character is", s[0], "and last character is", s[-1])
print("\nCharacters from index 2 to 6 are: ", s[2:7])
print("\nReverse of the given string is: ", s[::-1])
print("\nNumber of times 'e' letter appears is: ", s.count('e'))
print("\nUppercase of the given string is: ", s.upper())
```

2. **Given:**  
**s = "programming"**  

- **Create a dictionary that stores frequency of each character.**
- **Find the most frequent character.**
- **Remove duplicate characters (keep first occurrence).**
- **Print characters that appear only once.**
- **Check if the string is a palindrome.**


```python
s="programming"
dict={}
for i in s:
    dict[i]=dict.get(i, 0)+1
print("\nThe dictionary with the frequency is: ", dict)
max_char=max(dict, key=dict.get)
print("\nThe most frequent charcater is: ", max_char)
unique=""
for i in s:
    if i not in unique:
        unique+=i
print("\nUnique occurence is: ", unique)
single_char=[i for i in s if dict[i]==1]
print("\nUnique characters that appear only once are: ", single_char)
print("\nIs palindrome: ", s==s[::-1])
```

3. **Given:**  
**sentence = "python is powerful and python is easy"**  

- **Count total number of words.**
- **Count how many times the word "python" appears.**
- **Create a dictionary storing frequency of each word.**
- **Find the longest word.**
- **Replace all occurrences of "python" with "Python".**


```python
sentence="python is powerful and python is easy"
words=sentence.split()
print("\nTotal number of words are: ", len(words))
print("\nNumber of times word 'python' appears is: ", words.count("python"))
dict={}
for i in words:
    dict[i]=dict.get(i, 0)+1
print("\nFrequency of each word is: ", dict)
long_word=max(words, key=len)
new_sen=sentence.replace("python", "Python")
print("\nReplaced sentence is: ", new_sen)
```

4. **Given:**  
**s = "Cyber123Security456"**  

- **Extract only alphabets.**
- **Extract only digits.**
- **Count total digits.**
- **Replace digits with '*'.**
- **Check if the string contains both letters and numbers.**


```python
s="Cyber123Security456"
alpha="".join([ch for ch in s if ch.isalpha()])
print("\nOnly alphabets: ", alpha)
num="".join([ch for ch in s if ch.isdigit()])
print("\nOnly numbers: ", num)
```
