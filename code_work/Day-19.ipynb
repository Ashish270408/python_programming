1. **Create a function called divide(a, b).**  

- **It should return a / b.**
- **If b is 0, it should return "Division not possible".**
- **Call the function with:**
   - **Valid numbers**
   - **Zero as second argument**


```python
def divide(a, b):
    if b==0:
        return "Division not possible"
    return a/b
print(divide(4, 2))
print(divide(24, 0))
```

2. **Create a function called is_prime(n).**  

- **Return True if n is prime.**
- **Return False otherwise.**
- **Handle numbers less than 2 properly.**
- **Test it with multiple values.**


```python
def is_prime(n):
    if n<2:
        return False
    
    for i in range(2, n):
        if n%i==0:
            return False
    return True

print(is_prime(24))
print(is_prime(23))
print(is_prime(1))
print(is_prime(-124))
```

3. **Create a function called double_elements(lst).**  

- **It should return a new list where each element is doubled.**
- **Do NOT modify original list.**
- **Handle empty list.**
- **Use either loop or list comprehension.**


```python
def double_elements(lst):
    if len(lst)==0:
        return "List is empty"
    
    return [num*2 for num in lst]

l1=[2, 4, 5, 3, 9]
l2=[]
print(double_elements(l1))
print(double_elements(l2))
```

4. **Create a function called sum_all( * args).**

- **It should take any number of arguments.**
- **Return their sum.**
- **If no arguments are given, return 0.**
- **Test with different number of inputs.**

```python
def sum_all(*args):
    total=0
    for num in args:
        total+=num
    return total
print(sum_all(2, 3, 4, 5))
print(sum_all(-100, -34, 87))
```
