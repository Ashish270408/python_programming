1. **Given:**  
**s = "python is very powerful"**

- **Reverse the order of words.**
- **Reverse each word individually.**
- **Capitalize the first letter of each word.**
- **Count total vowels.**
- **Find the longest word.**


```python
s="python is very powerful"
words=s.split()
rev_words=words[::-1]
rev_s=" ".join(rev_words)
print("\nReverse order of words of given string is: \n", rev_s)
rev_w=[]
for i in words:
    rev_w.append(i[::-1])
rev_str=" ".join(rev_w)
print("\nString with each word reversed individually is: \n", rev_str)
print("\nString with first letter of each word capitalized is: \n", s.title())
count=0
for i in s:
    if i.lower() in ['a', 'e', 'i', 'o', 'u']:
        count+=1
print("\nTotal vowels in given string: ", count)
max_word=max(words, key=len)
print("\nLongest word in given string is: ", max_word)
```

2. **Given:**  
**s = "aaabbcdddde"**  

- **Find frequency of each character.**
- **Find first repeating character.**
- **Find first non-repeating character.**
- **Compress the string (a3b2c1d4e1).**
- **Remove all duplicate characters.**


```python
s="aaabbcdddde"
freq={}
for i in s:
    freq[i]=freq.get(i, 0)+1
print("\nFreqency of given string is: \n", freq)
for i in s:
    if freq[i]>1:
        print("\nFirst repeating character is: ", i)
        break
for i in s:
    if freq[i]==1:
        print("\nFirst non-repeating character is: ", i)
        
compress=""
for i in freq:
    compress+=i
    compress+=str(freq[i])
print("\nCompressed string is: ", compress)
unique=""
for i in s:
    if i not in unique:
        unique+=i
print("\nString with no duplicates is: ", s)
```

3. **Given:**  
**nums = [5, 3, 8, 3, 9, 1, 5]**  

- **Remove duplicates (keep order).**
- **Find third largest element without sorting.**
- **Rotate list to the right by 2 positions.**
- **Find all pairs whose sum = 10.**
- **Check if list is sorted.**


```python
nums=[5, 3, 8, 3, 9, 1, 5]
unique=[]
for i in nums:
    if i not in unique:
        unique.append(i)
print("\nList with only unique values is: ", unique)
largest=second=third=float('-inf')
for i in  nums:
    if i>largest:
        third=second
        second=largest
        largest=i
    elif i>second and i!=largest:
        third=second
        second=i
    elif i>third and i!=second:
        third=i
print("\nThird largest element in given list is: ", third)
rotate=nums[-2:]+nums[:-2]
print("\nRotated list is: ", rotate)
target=10
pairs=[]
for i in nums:
    complement=target-i
    if complement in nums:
        pairs.append((i, complement))
print("\nAll such pairs are: ", pairs)
print("\nIs sorted: ", nums==nums[::-1])
```
