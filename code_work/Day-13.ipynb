1. **Create a NumPy array from 1 to 24.**  
**Reshape it into a 4x6 matrix.**  
**Then:**  
- **Print the matrix.**
- **Print the mean of each column.**
- **Print the mean of each row.**
- **Print the shape after applying mean(axis=0).**
- **Print the shape after applying mean(axis=1).**


```python
a=np.arange(1, 25)
b=a.reshape(4, 6)
print("\nThe matrix is: \n", b)
row=b.mean(axis=1)
col=b.mean(axis=0)
print("\nThe mean of each column is: \n", col)
print("\nThe mean of each row is: \n", row)
print("\nThe shape of mean(axis=0) is: \n", col.shape)
print("\nThe shape of mean(axis=1) is: \n", row.shape)
```

```python
modify=b.copy()
modify=modify+10
r=np.arange(1, 7)
modify=modify+r
print(modify)
```

3. **Using the same 4x6 matrix:**
  
- **Find all elements divisible by 5.**
- **Print their positions (row, column).**
- **Count how many such elements exist.**
- **Replace those elements with -99.**
- **Print the modified matrix.**


```python
mask=b%5==0
print("\nElements divisble by 5 are: \n", b[mask])
positions=np.where(mask)
coords=list(zip(positions[0], positions[1]))
print("\nThe coordinates of such elements are: \n", coords)
count=np.sum(mask)
print("\nNumber of such elements are: ", count)
b[mask]=-99
print("\nThe modified matrix is: \n", b)
```
