1. **Create a function called find_max_value( * args).**  

- **It should accept any number of arguments.**
- **Return the largest value.**
- **If no arguments are passed, return None.**
- **Do NOT use max().**


```python
def max_value(*args):
    if len(args)==0:
        return None
    largest=args[0]
    for num in args:
        if num>largest:
            largest=num
            
    return largest
print(max_value(2, 3, 4, 1, 9, 5, 0))
print(max_value())
```

2. **Create a function called count_types( * args).**  

- **It should accept multiple arguments.**
- **Return how many are:**
   - **integers**
   - **strings**
   - **floats**
- **Return the result as a dictionary.**


```python
def count_types(*args):
    types={"int":0, "str":0, "float":0}
    for num in args:
        if isinstance(num, int):
            types["int"]+=1
        elif isinstance(num, str):
            types["str"]+=1
        elif isinstance(num, float):
            types["float"]+=1
    return types
print(count_types(1, 27.4, "Ashish", "hello9", 56.3))
```

3. **Create a function called flatten(lst).**  
**Input example:**  
**[1, 2, [3, 4], 5, [6, 7]]**  
**Output:**  
**[1, 2, 3, 4, 5, 6, 7]**  
**Assume only one level of nesting.**


```python
def flatten(lst):
    result=[]
    for num in lst:
        if isinstance(num, list):
            for values in num:
                result.append(values)
        else:
            result.append(num)
            
    return result

print(flatten([1, 2, [3, 4], 5, [6, 7]]))
```

4. **Create a function called analyze_list(lst).**
- **It should return:**
   - **Sum of elements**
   - **Average**
   - **Maximum**
   - **Minimum**
- **Handle empty list (return None for all).**
- **Do NOT use built-in sum(), max(), min().**


```python
def analyze_list(lst):
    sum=0
    average=0
    maximum=lst[0]
    minimum=lst[0]
    for num in lst:
        if num>maximum:
            maximum=num
        elif num<minimum:
            minimum=num
        sum+=num
    average=sum/len(lst)
    
    return sum, average, maximum, minimum

total, avg, max_value, min_value = analyze_list([2, 4, 1, 5, 9, 6, 8, 10])
print(f"Sum: {total}, Average: {avg}, Max Value: {max_value}, Min Value: {min_value}")
```
